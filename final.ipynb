## Importar librerías

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Leer imagenes de los canales R, G y B

```python
IMAGES_PATH = "images/"

R_channel = cv.imread(f"{IMAGES_PATH}red1.jpeg", cv.IMREAD_UNCHANGED)
R_channel[:,:,1]=0
R_channel[:,:,2]=0

B_channel = cv.imread(f"{IMAGES_PATH}blue1.jpeg", cv.IMREAD_UNCHANGED)
B_channel[:,:,0]=0
B_channel[:,:,1]=0

G_channel = cv.imread(f"{IMAGES_PATH}green1.jpeg", cv.IMREAD_UNCHANGED)
G_channel[:,:,0]=0
G_channel[:,:,2]=0
```

## Mostrando imágenes de los canales R, G y B

```python
plt.subplots(figsize=(12, 3))

# Canal Red
plt.subplot(1,3,1)
plt.imshow(R_channel)
plt.title("Red Channel")
plt.axis("off")

# Canal Blue
plt.subplot(1,3,2)
plt.imshow(B_channel)
plt.title("Blue Channel")
plt.axis("off")

# Canal Green
plt.subplot(1,3,3)
plt.imshow(G_channel)
plt.title("Green Channel")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## Obteniendo el tamaño de la imagen

```python
img_size = R_channel.shape
```

## Combinando las imágenes de los canales R, G y B para obtener una sola

```python
combinated_img = np.zeros_like(R_channel, dtype=np.uint8)
combinated_img = cv.add(combinated_img, R_channel)
combinated_img = cv.add(combinated_img, G_channel)
combinated_img = cv.add(combinated_img, B_channel)

plt.imshow(combinated_img)
plt.title("Combinated Image")
plt.axis("off")
plt.tight_layout()
plt.show()
```

## Comparando la imagen combinada con una tomada a color con nuestra cámara

```python
# Imagen original

original_img = cv.imread(f"{IMAGES_PATH}original.jpeg")
original_img = cv.cvtColor(original_img, cv.COLOR_BGR2RGB)

plt.figure(figsize=(15, 12))

# Imagen original
plt.subplot(1,3,2)
plt.imshow(original_img)
plt.title("Imagen original")
plt.axis("off")

# Imagen combinada
plt.subplot(1,3,3)
plt.imshow(combinated_img)
plt.title("Imagen combinada")
plt.axis("off")

plt.tight_layout()
plt.show()
```
